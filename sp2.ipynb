```python
# SOFTWARE PROJECT 2
# -----------------------

class Node:
    def __init__(self, name):
        self.name = name
        self.neighbors = []  # a list containing the neighbors of Node object; for MS1
        self.Wtdneighbors = {} # a dictionary containing the neighbors of Node object; for MS2

    def add_neighbor(self, node):
        # a Helper function that adds a neighbor to a Node object
        if node not in self.neighbors:
            self.neighbors.append(node)

class Graph:
    def __init__(self):
        self.nodes = []  # All nodes are stored here

    def get_node(self, name):
        # Searches for the node with matching name. Returns None if not found
        for node in self.nodes:
            if node.name == name:
                return node
        return None

    def add_node(self, name):
        # Creates a Node object and adds it to the Graph object
        # Checks if node is already existing
        existing = self.get_node(name)
        if (existing):
            return existing
        # Creates a new node if node is not found
        new_node = Node(name)
        self.nodes.append(new_node)
        return new_node

    def add_edge(self, src, dst):

# MILESTONE 1
        start = self.add_node(src)
        end = self.add_node(dst)

        if end not in start.neighbors: # to avoid duplicate neighbors
            start.neighbors.append(end)
            end.neighbors.append(start)

    def __str__(self):
        # Prints the graph in Adjacency List format
        out = ""
        for node in self.nodes:
            out += f" {node.name} -> ["
            for neighbor in node.neighbors:
                out += f" {neighbor.name}, "
            out += f"]\n"
        return out

# MILESTONE 2

class WeightedGraph(Graph):
    def __init__(self):
        super().__init__() #inheritance from parent class
        self.total = 0

    def add_edge(self, src, dst, cst=1, dir=False): # "=(values)" initializes cost and directionality defaults

        start = self.add_node(src)
        end = self.add_node(dst)

        if end not in start.Wtdneighbors: # to avoid duplicate neighbors
            start.Wtdneighbors[end] = (end.name, cst)

            if not dir and (start not in end.Wtdneighbors): #to assign different cost if directional
                end.Wtdneighbors[start] = (start.name, cst)

# MILESTONE 3
    def traverse(self, start):

        current_node = self.get_node(start)

        '''
        things to do tomorrow:
        1. available neighbors
        2. showing current node before traversal
        3. showing current node after traversal and cost in progress
        '''

        while True:

            print(f"Available neighbors: ")
            choice = input("Enter the next node: ")
            new_node = self.get_node(choice)

            if not new_node:
                print("This node does not exist. Enter another.")
            else:
                if new_node in current_node.Wtdneighbors.keys():
                    cost = current_node.Wtdneighbors[new_node][1]
                    self.total = self.total + cost
                    if new_node.name == "End":
                        break
                    else:
                        current_node = new_node
                else:
                    print("Not a neighboring node. Enter another.")

        print(self.total)

    def __str__(self):
    # Prints the graph in Adjacency List format
        out = ""
        for node in self.nodes:
            out += f" {node.name} -> ["
            for neighbor in node.Wtdneighbors:
                out += f" {node.Wtdneighbors[neighbor]}" #utilizes "Wtdneighbors" dictionary to access node and cost values
            out += f"]\n"

        return out


# MS 1: Test Inputs
if __name__ == "__main__":
    g = Graph()

    g.add_node("Start")
    g.add_node("A")
    g.add_node("B")
    g.add_node("C")
    g.add_node("End")

    g.add_edge("Start", "A")
    g.add_edge("A", "B")
    g.add_edge("B", "C")
    g.add_edge("A", "C")
    g.add_edge("C", "A")
    g.add_edge("C", "End")

    print(g)

# MS 2 & 3: Test Inputs
if __name__ == "__main__":
    g = WeightedGraph()

    g.add_node("Start")
    g.add_node("A")
    g.add_node("B")
    g.add_node("C")
    g.add_node("End")

    g.add_edge("Start", "A")
    g.add_edge("A", "B")
    g.add_edge("B", "C", 2)
    g.add_edge("A", "C", 67, True)
    g.add_edge("C", "A", 76, True)
    g.add_edge("C", "End")

    g.traverse("Start")

    print(g)
```
